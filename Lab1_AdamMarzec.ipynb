```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x) # 3th curve 2x
plt.figure(figsize=(6,4))
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.plot(x, y3, label="sin(2x)") #3th curve label
plt.title("Trigonometric functions")
plt.xlabel("x (radians)")
plt.ylabel("y")
plt.ylim(-1.2, 1.2) # y limit
plt.xlim(0, np.pi) # x limit
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

print("=== Simple Scientific Calculator ===")
print("Options: add, subtract, multiply, divide, power, sqrt")

while(True):
  choice = input("Choose an operation: ")
  if choice == "add":
    a = float(input("a = "))
    b = float(input("b = "))
    print("Result: ", a + b)
  elif choice == "subtract":
    a = float(input("a = "))
    b = float(input("b = "))
    print("Result: ", a - b)
  elif choice == "multiply":
    a = float(input("a = "))
    b = float(input("b = "))
    print("Result: ", a * b)
  elif choice == "divide":
    a = float(input("a = "))
    b = float(input("b = "))
    print("Result: ", a / b)
  elif choice == "power":
    a = float(input("Base = "))
    b = float(input("Exponent = "))
    print("Result: ", a ** b)
  elif choice == "sqrt":
    a = float(input("Number = "))
    print("Result: ", math.sqrt(a))
  elif choice == "sin":
    a = float(input("Number = "))
    print("Result: ", math.sin(a))
  elif choice == "cos":
    a = float(input("Number = "))
    print("Result: ", math.cos(a))
  elif choice == "tan":
    a = float(input("Number = "))
    print("Result: ", math.tan(a))
  else:
    print("Operation not recognized")

  if input("Do you want to continue? (y/n): ") != "y":
    break
```
